```python
import matplotlib.pyplot as plt
import numpy as np
import pickle

def load_results(filename):
    with open(filename, 'rb') as inp:
        results = pickle.load(inp)
    return results


plt.rc('text', usetex=True)
plt.rc('font', family='Times New Roman')
```

```python
update_mode = 'newUE'
csi_mode = 'imperfect'
nbrOfSetups = 1
nbrOfNewUEs = 6

file_name = (f'./ToyModelsSE_{update_mode}_CSI_{csi_mode}_NbrSetups_{nbrOfSetups}'
             f'_NbrUEs_{nbrOfNewUEs}.pkl')

SEs_newUE = load_results(file_name)
UE0_newUE = SEs_newUE[:, 0]
UE1_newUE = SEs_newUE[:, 1]
UE2_newUE = SEs_newUE[:, 2]
UE3_newUE = SEs_newUE[:, 3]
UE4_newUE = SEs_newUE[:, 4]
UE5_newUE = SEs_newUE[:, 5]


update_mode = 'allUEs'
file_name = (f'./ToyModelsSE_{update_mode}_CSI_{csi_mode}_NbrSetups_{nbrOfSetups}'
             f'_NbrUEs_{nbrOfNewUEs}.pkl')

SEs_allUEs = load_results(file_name)
UE0_allUEs = SEs_allUEs[:, 0]
UE1_allUEs = SEs_allUEs[:, 1]
UE2_allUEs = SEs_allUEs[:, 2]
UE3_allUEs = SEs_allUEs[:, 3]
UE4_allUEs = SEs_allUEs[:, 4]
UE5_allUEs = SEs_allUEs[:, 5]


xAxis = ['t0', 't1', 't2', 't3', 't4', 't5']

fig = plt.figure()
ax = fig.add_subplot(111)

plt.grid(visible = True, linestyle='--')
ax.set_ylabel(r'SE [bit/s/Hz]', size=15)
plt.xlabel(r'time instant', size=15)
plt.plot(xAxis, UE0_allUEs,'m-*', label = r'UE0_allUEs')
plt.plot(xAxis, UE0_newUE,'m--*', label = r'UE0_newUE')
plt.plot(xAxis, UE1_allUEs,'g-*', markersize = 9, label = r'UE1_allUEs')
plt.plot(xAxis, UE1_newUE,'g--*', markersize = 9, label = r'UE1_newUE')
plt.plot(xAxis, UE2_allUEs,'c-d', markersize = 9, label = r'UE2_allUEs')
plt.plot(xAxis, UE2_newUE,'c--d', markersize = 9, label = r'UE2_newUE')
plt.plot(xAxis, UE3_allUEs,'r-p', label = r'UE3_allUEs')
plt.plot(xAxis, UE3_newUE,'r--p', label = r'UE3_newUE')
plt.plot(xAxis, UE4_allUEs,'b-s', label = r'UE4_allUEs')
plt.plot(xAxis, UE4_newUE,'b--s', label = r'UE4_newUE')
plt.plot(xAxis, UE5_allUEs,'y-^', label = r'UE5_allUEs')
plt.plot(xAxis, UE5_newUE,'y--^', label = r'UE5_newUE')

plt.legend()
plt.draw()
plt.show()

image_format = 'jpeg' # e.g .png, .svg, etc.
image_name = 'SE_ToyModel_allUEsVSnewUE.jpeg'
fig.savefig(image_name, format=image_format, dpi=400)
```
