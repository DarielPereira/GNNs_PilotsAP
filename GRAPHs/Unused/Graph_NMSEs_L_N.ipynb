```python
import matplotlib.pyplot as plt
import numpy as np
from functionsUtils import load_results

plt.rc('text', usetex=True)
plt.rc('font', family='Times New Roman')
```

```python
results = load_results('./VARIABLES_SAVED/SR_NMSE_L_K_100_NbrSetps_30_extended.pkl')

Kmeans_SRs = results['Kmeans_locations']['sum_rates']
Kmeans_NMSEs = results['Kmeans_locations']['NMSEs']
Kfootprints_SRs = results['Kfootprints']['sum_rates']
Kfootprints_NMSEs = results['Kfootprints']['NMSEs']
DCC_SRs = results['DCC']['sum_rates']
DCC_NMSEs = results['DCC']['NMSEs']
DCPA_SRs = results['DCPA']['sum_rates']
DCPA_NMSEs = results['DCPA']['NMSEs']
balanced_random_SRs = results['balanced_random']['sum_rates']
balanced_random_NMSEs = results['balanced_random']['NMSEs']
random_SRs = results['random']['sum_rates']
random_NMSEs = results['random']['NMSEs']


xAxis = ['(4, 100)', '(9, 36)', '(16, 25)', '(36, 16)',' (64, 9)', '(100, 4)', '(400, 1)']

fig = plt.figure()
ax = fig.add_subplot(111)

plt.grid(visible = True, linestyle='--')
ax.set_ylabel(r'sum-rate [bit/s/Hz]', size=15)
plt.xlabel(r'(\textit{L}, \textit{N})', size=15)
plt.plot(xAxis, Kfootprints_SRs,'c-d', markersize = 9, label = r'K-footprints+BFPA')
# plt.plot(xAxis, Kmeans_SRs,'b--*', markersize = 9, label = r'K-means+BFPA')
plt.plot(xAxis, DCC_SRs,'g--*', label = r'DCC')
plt.plot(xAxis, DCPA_SRs,'k--p', label = r'DCPA')
plt.plot(xAxis, balanced_random_SRs,'r--p', label = r'balanced random')
plt.plot(xAxis, random_SRs,'y--p', label = r'random')

# plt.xlim(18,102)

plt.legend(fontsize=10)


plt.draw()
plt.show()

image_format = 'pdf' # e.g .png, .svg, etc.
image_name = 'SR_L_N.pdf'
fig.savefig(image_name, format=image_format, dpi=400)
```

```python
results = load_results('./VARIABLES_SAVED/SR_NMSE_L_K_100_NbrSetps_30_extended.pkl')

Kmeans_NMSEs = results['Kmeans_locations']['NMSEs']
Kfootprints_NMSEs = results['Kfootprints']['NMSEs']
DCC_NMSEs = results['DCC']['NMSEs']
DCPA_NMSEs = results['DCPA']['NMSEs']
balanced_random_NMSEs = results['balanced_random']['NMSEs']
random_NMSEs = results['random']['NMSEs']


xAxis = ['(4, 100)', '(9, 36)', '(16, 25)', '(36, 16)',' (64, 9)', '(100, 4)', '(400, 1)']

fig = plt.figure()
ax = fig.add_subplot(111)

plt.grid(visible = True, linestyle='--')
ax.set_ylabel(r'sum-NMSE', size=15)
plt.xlabel(r'(\textit{L}, \textit{N})', size=15)

plt.plot(xAxis, Kfootprints_NMSEs,'c-d', markersize = 9, label = r'C-footprints+CF-BFPA')
plt.plot(xAxis, Kmeans_NMSEs,'g--*', markersize = 9, label = r'K-means+CF-BFPA')
plt.plot(xAxis, DCC_NMSEs,'m--*', label = r'BPA')
plt.plot(xAxis, DCPA_NMSEs,'r--1', label = r'DCPA')
plt.plot(xAxis, balanced_random_NMSEs,'y--<', label = r'balanced random')
plt.plot(xAxis, random_NMSEs,'b-->', label = r'fully random')

# plt.xlim(18,102)

plt.legend(fontsize=10)


plt.draw()
plt.show()

image_format = 'pdf' # e.g .png, .svg, etc.
image_name = 'NMSE_L_N.pdf'
fig.savefig(image_name, format=image_format, dpi=400)
```
