```python
import matplotlib.pyplot as plt
import numpy as np

plt.rc('text', usetex=True)
plt.rc('font', family='Times New Roman')
image_format = 'jpeg'
```

```python
variables = np.load('./VARIABLES_SAVED/ClusteringOrthogonality_0.08726646259971647.npz')
Kbeams = variables['K_beams']
Kmeans = variables['Kmeans_basic_positions']

xAxis = [20, 30, 40, 50, 60, 70, 80, 90, 100]

fig, ax = plt.subplots()
plt.grid(visible = True, linestyle='--')
plt.ylabel(r'orthogonality', size=15)
plt.xlabel(r'\textit{K}', size=15)
plt.plot(xAxis, Kbeams[0],'b-*', markersize = 9, label = r'K-beams (intra-cluster orthogonality)')
plt.plot(xAxis, Kbeams[1],'b--*', markersize = 9, label = r'K-beams (inter-cluster orthogonality)')
plt.plot(xAxis, Kmeans[1],'g--^', label = r'K-means (inter-cluster orthogonality)')
plt.plot(xAxis, Kmeans[0],'g-^', label = r'K-means (intra-cluster orthogonality)')
plt.xlim(18,102)
plt.ylim(-0.2,11)

plt.legend(fontsize=10)


plt.draw()
plt.show()

image_format = 'pdf' # e.g .png, .svg, etc.
image_name = 'ClusteringOrthogonality.pdf'
fig.savefig(image_name, format=image_format, dpi=400)
```
