```python
import matplotlib.pyplot as plt
import numpy as np

plt.rc('text', usetex=True)
plt.rc('font', family='Times New Roman')
image_format = 'jpeg'
```

```python

M_ = 100
K_ = 25
KB_worstcase = []
KB_parallel = []
SGPS = []

for i in range(1,6):
    M = M_*i
    K = K_*i


    KB_worstcase.append((K**2+M))
    KB_parallel.append((K+M))
    SGPS.append((K**3))
    
xAxis = ['1x','2x','3x','4x', '5x']

fig, ax = plt.subplots()
plt.grid(visible = True, linestyle='--')
plt.ylabel(r'matrix multiplications [log\textsubscript{10} scale]', size=15)
plt.xlabel(r'setups', size=15)
plt.plot(xAxis, np.log10(SGPS),'m-o', label = r'SGPS')
plt.plot(xAxis, np.log10(KB_worstcase),'b-*', markersize = 9, label = r'K-beams+BFPA (worst-case)')
plt.plot(xAxis, np.log10(KB_parallel), 'b--*', markersize = 9,  label = r'K-beams+BFPA (practical)')

plt.legend(fontsize=10)


plt.draw()
plt.show()

image_format = 'pdf' # e.g .png, .svg, etc.
image_name = 'CC_KbeamsBFPA.pdf'
fig.savefig(image_name, format=image_format, dpi=400)
```
